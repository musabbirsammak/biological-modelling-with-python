The HH model describes the time evolution of the membrane potential $V$ and the gating variables that control ion flow. The model is defined by a set of four coupled ordinary differential equations (ODEs):

1. **Membrane Potential:** $C_m \frac{dV}{dt} = - \bar{g}_{Na}m^3h(V - E_{Na}) - \bar{g}_{K}n^4(V - E_{K}) - g_L(V - E_L) + I_{ext}$    
2. **Potassium Activation ($n$):** $\frac{dn}{dt} = \alpha_n(V)(1-n) - \beta_n(V)n$
3. **Sodium Activation ($m$):** $\frac{dm}{dt} = \alpha_m(V)(1-m) - \beta_m(V)m$
4. **Sodium Inactivation ($h$):** $\frac{dh}{dt} = \alpha_h(V)(1-h) - \beta_h(V)h$

| **Symbol**    | **Biological Meaning** | **Physical Units** | **Placement Logic**                                               |
| ------------- | ---------------------- | ------------------ | ----------------------------------------------------------------- |
| $V$           | Membrane Potential     | $mV$               | The state variable we are tracking over time.                     |
| $C_m$         | Membrane Capacitance   | $\mu F/cm^2$       | Represents the bilayer's ability to store charge.                 |
| $g_{Na}, g_K$ | Conductances           | $mS/cm^2$          | The "ease" with which ions flow; $1/R$ in electrical terms.       |
| $E_{Na}, E_K$ | Equilibrium Potentials | $mV$               | The "battery" voltage where net ion flow is zero.                 |
| $n, m, h$     | Gating Variables       | Dimensionless      | Probabilities (0 to 1) that specific channel subunits are "open". |
| $I_{ext}$     | External Current       | $\mu A/cm^2$       | The stimulus applied to the neuron.                               |

Hodgkin and Huxley arrived at these equations by combining experimental data from the Squid Giant Axon with the principles of electrical engineering. Using a voltage clamp—a then-revolutionary technique—they held the membrane potential at fixed levels to isolate and measure the individual currents flowing through the membrane. By replacing the external sodium with choline to block sodium current, they could mathematically subtract one current from the total to isolate the potassium current. They observed that these currents did not respond instantly to voltage changes but followed a specific time-course: the potassium current rose with a sigmoid "lag," while the sodium current rose quickly and then spontaneously decayed.

To translate these physical observations into math, they treated the membrane as a parallel electrical circuit. They hypothesized that the gates controlling the ions were physical particles that had to move to an open position. By fitting the experimental curves of the potassium conductance, they found that the data best matched a variable ($n$) raised to the fourth power, suggesting four identical subunits must be open. For sodium, the fast rise followed by a slow drop required two separate variables—one for activation ($m$) and one for inactivation ($h$). They then calculated the rate constants ($\alpha$ and $\beta$) for these variables by measuring how the speed of opening and closing changed at different clamped voltages, eventually assembling these empirical fits into the four coupled differential equations you have in your notes.

For this system to work, we make the following assumptions:

- **The Membrane as a Capacitor:** The phospholipid bilayer acts as an insulator separating two conducting media, effectively storing charge.
- **Independent Ion Channels:** Current is carried by specific ions ($Na^+$, $K^+$, and a general "leakage") through independent, voltage-gated channels.
- **Gating Particles:** Ion conductances are governed by "gates" that move between open and closed states based on first-order kinetics.

In the Hodgkin-Huxley model, the neuron is treated as an equivalent electrical circuit where every biological structure has a specific electronic counterpart. 

1. The **capacitor** represents the **phospholipid bilayer** of the cell membrane. A capacitor consists of two conducting plates separated by an insulator. In a neuron, the salty fluids inside (cytoplasm) and outside the cell are the "conducting plates," and the fatty membrane itself—which is impermeable to ions—is the "insulator". Because the membrane is so thin (~5 nm), it can store opposite electrical charges on its surfaces. This "stored charge" creates the membrane potential ($V$).
2. An **ionic branch** represents a specific pathway (an ion channel) through which a particular type of ion can cross the membrane. Each branch consists of two parts: **The Conductance ($g$)** represents the **ion channels**. In the circuit, this is a resistor (or more accurately, a conductor). For Sodium ($Na^+$) and Potassium ($K^+$), these are **voltage-gated**, meaning the "resistance" changes depending on the membrane's voltage.
3. **The Electromotive Force ($E$)** represents the **reversal potential** (or Nernst potential) for that specific ion. In the circuit, this is a **battery**. Because there are different concentrations of ions inside vs. outside, each ion has a "natural" pressure to move in a certain direction. The battery represents the "voltage" provided by those concentration gradients.

Kirchhoff's Current Law states that the algebraic sum of all currents entering and leaving a junction must equal zero, meaning that charge is conserved within the circuit. 

Since $I = \frac{dq}{dt}$ and $q = C_m V$, the term $C_m \frac{dV}{dt}$ represents the physical flow of charge (current) required to change the voltage across the membrane's capacitance. 

Now, $I_{ext} = C_m \frac{dV}{dt} + I_{ion}$, where the total applied current must be partitioned into either charging the membrane capacitor or flowing through the ionic resistors. When we write the equation $C_m \frac{dV}{dt} = -I_{ion} + I_{ext}$:

1. **Current enters the system** ($I_{ext}$), such as a probe injecting charge into the cell.
2. **The Capacitor charges first:** Initially, the charge sticks to the membrane ($C_m \frac{dV}{dt}$), causing the voltage to rise.
3. **The Gates open:** As the voltage rises, the "resistors" in the ionic branches (the $m$ and $n$ gates) start to change their values.
4. **Ions flow through the branches:** Sodium rushes in (depolarization), and later Potassium rushes out (repolarization), following the "batteries" ($E_{Na}$, $E_K$) provided by nature.

The ionic current is the sum of individual ion flows:
$$I_{ion} = I_{Na} + I_K + I_L$$
For each ion, the current depends on the driving force (the difference between the actual membrane potential and that ion's equilibrium potential).
$$I_i = g_i(V - E_i)$$
Substituting this into our balance equation, we get the fundamental HH voltage equation:
$$C_m \frac{dV}{dt} = -[g_{Na}(V - E_{Na}) + g_K(V - E_K) + g_L(V - E_L)] + I_{ext}$$
The conductances are not constant; they change as "gates" open or close.

- **Potassium ($K$):** Observed to have a "delayed rectifier" behavior. HH modeled this as 4 independent $n$-gates that must all be open : $g_K = \bar{g}_K n^4$.
- **Sodium ($Na$):** Shows fast activation and slower inactivation. This requires two variables: $m$ (activation) and $h$ (inactivation). Specifically, 3 $m$-gates and 1 $h$-gate must be open : $g_{Na} = \bar{g}_{Na} m^3 h$.

Each gate ($n, m, h$) follows first-order kinetics between an **Open** and **Closed** state:
$$\text{Closed} \xrightarrow{\alpha(V)} \text{Open} \quad ; \quad \text{Open} \xrightarrow{\beta(V)} \text{Closed}$$
The rate of change of the open fraction is (Rate Open $\times$ Fraction Closed) - (Rate Close $\times$ Fraction Open):
$$\frac{dn}{dt} = \alpha_n(V)(1-n) - \beta_n(V)n$$
**Step A: Distribute $\alpha_n$**
$$\frac{dn}{dt} = \alpha_n - \alpha_n n - \beta_n n$$
**Step B: Factor out the $-n$**
$$\frac{dn}{dt} = \alpha_n - n(\alpha_n + \beta_n)$$
**Step C: Force the $(\alpha_n + \beta_n)$ term out of the whole expression**. To make the right side look like a fraction, we multiply and divide the $\alpha_n$ term by $(\alpha_n + \beta_n)$: $$\frac{dn}{dt} = (\alpha_n + \beta_n) \left[ \frac{\alpha_n}{\alpha_n + \beta_n} - n \right]$$**Step D: Define the Biological Constants** Now we define our two "Steady State" and "Time" anchors:

1. **The Steady State ($n_\infty$)**: The value $n$ reaches if $V$ is held constant forever. Mathematically, as $t \to \infty$, the derivative $\frac{dn}{dt}$ becomes zero, and solving the first-order kinetic equation $\alpha_n(1-n) - \beta_n n = 0$ for $n$ yields this specific ratio as the only stable equilibrium point. $$n_\infty(V) = \frac{\alpha_n(V)}{\alpha_n(V) + \beta_n(V)}$$
2. **The Time Constant ($\tau_n$)**: How quickly the gate reacts to a change. $$\tau_n(V) = \frac{1}{\alpha_n(V) + \beta_n(V)}$$In a two-state system, **$\alpha$** is the rate of "closed-to-open" jumps and **$\beta$** is the rate of "open-to-closed" jumps. The **total rate of transition** ($\alpha + \beta$) tells you how much "flipping" is happening in total per millisecond. If this total rate is high, the system is "fast" and reaches equilibrium quickly, meaning the **Time Constant** ($\tau = 1 / \text{total rate}$) is small. If the total rate is low, the system is "sluggish" and takes forever to catch up, meaning $\tau$ is large.

>	This equation says that the speed at which the gate moves ($\frac{dn}{dt}$) is directly proportional to how far it currently is from its target ($n_\infty - n$), with the time constant ($\tau_n$) acting as the "brake" that determines how long that adjustment takes.

**Step E: Final Substitution**: Substitute $n_\infty$ and $1/\tau_n$ back into the equation from Step C:
$$\frac{dn}{dt} = \frac{1}{\tau_n} [n_\infty - n]$$
Which gives us the final "standard" form:
$$\tau_n(V) \frac{dn}{dt} = n_\infty(V) - n$$
---

The Hodgkin-Huxley system is defined by four state variables: $V, m, n,$ and $h$. Because these equations are non-linear and coupled, their stability is determined by the interplay of fast positive feedback and slower negative feedback.

A steady state occurs when the rate of change for all variables is zero ($\frac{dV}{dt} = 0, \frac{dm}{dt} = 0, \frac{dn}{dt} = 0, \frac{dh}{dt} = 0$).

- **Resting Potential ($V_{rest}$):** In the absence of external current ($I_{ext} = 0$), the system settles at a stable fixed point. At this voltage (typically around $-60\text{ mV}$ to $-70\text{ mV}$), the inward sodium current and outward potassium current exactly cancel each other out.
- **Gating Equilibria:** At any constant voltage $V$, the gating variables $m, n,$ and $h$ asymptotically approach their steady-state values $m_\infty(V), n_\infty(V),$ and $h_\infty(V)$.

To prove stability formally, we would look at the **Jacobian matrix ($J$)** of the system evaluated at $V_{rest}$:
$$J = \begin{pmatrix} \frac{\partial \dot{V}}{\partial V} & \frac{\partial \dot{V}}{\partial m} & \frac{\partial \dot{V}}{\partial h} & \frac{\partial \dot{V}}{\partial n} \\ \vdots & \vdots & \vdots & \vdots \end{pmatrix}$$

When we look at $J_{1,j} = \frac{\partial \dot{V}}{\partial x_j}$, we are measuring the "feedback" the neuron gives itself:

- **Self-Damping ($\frac{\partial \dot{V}}{\partial V}$):** This entry includes terms like $-g_L / C_m$. Mathematically, this is negative, meaning if the voltage increases, the leak current increases to pull it back down. This is the **restorative force** of the "sink."
- **The Accelerator ($\frac{\partial \dot{V}}{\partial m}$):** We differentiate the sodium current $- \bar{g}_{Na}m^3h(V - E_{Na})$ with respect to $m$. This gives a **positive** value. It tells us that increasing the sodium "openness" increases the rate of voltage rise. This is the **positive feedback** that causes the spike. The derivative $\frac{\partial \dot{V}}{\partial m}$ is positive because the negative sign in the equation $C_m \dot{V} = - \bar{g}_{Na} m^3 h (V - E_{Na})$ is multiplied by a **negative driving force** $(V - E_{Na})$—since $V$ is roughly $-60\text{ mV}$ and $E_{Na}$ is $+50\text{ mV}$ —meaning that increasing $m$ mathematically increases $\dot{V}$, proving the existence of an unstable, self-amplifying loop rather than a self-correcting one.
- **The Brakes ($\frac{\partial \dot{V}}{\partial n}$):** We differentiate the potassium current with respect to $n$. This is **negative**. Increasing potassium "openness" works against the voltage rise.

We don't just "look" at the matrix; we solve for its characteristic behavior. The eigenvalues $\lambda$ are the solutions to $\det(J - \lambda I) = 0$.

1. **Stable Rest:** At the resting potential, the "Restorative" and "Brake" terms dominate. The math results in eigenvalues with **negative real parts**. This means any perturbation decays like $e^{-|\lambda|t}$.
2. **The Instability (Hopf Bifurcation):** As $I_{ext}$ increases, the neuron depolarizes. At higher voltages, the "Accelerator" term ($\frac{\partial \dot{V}}{\partial m}$) grows much faster than the others.
    - Eventually, the real parts of a pair of complex eigenvalues cross from negative to **positive**.
    - The "sink" (stable point) disappears and is replaced by a **Limit Cycle**.

The interplay of Na-K transport causes the spike dynamics. Sodium conductance is fast and transient. Potassium is delayed and long-lasting (it does not deactivate, but remains open for as long as the membrane potential remains depolarized). Their joint action results in a brief and strong depolarization event for Vm followed by hyperpolarization.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
C_m  = 1.0      # Membrane Capacitance (uF/cm^2)
g_Na = 120.0    # Max Sodium Conductance (mS/cm^2)
g_K  = 36.0     # Max Potassium Conductance (mS/cm^2)
g_L  = 0.3      # Leak Conductance (mS/cm^2)
E_Na = 50.0     # Sodium Equilibrium Potential (mV)
E_K  = -77.0    # Potassium Equilibrium Potential (mV)
E_L  = -54.387  # Leak Equilibrium Potential (mV)
```

```python
# These are the empirical curve fits from the 1952 Hodgkin-Huxley paper.
# We add offsets (like +65.0) because HH defined Resting Potential as 0 mV, 
# but modern neuroscience defines it as -65 mV (so we shift the axis to match).

# Boltzmann Function (alpha_m): Creates the Threshold.
# At low V, this rate is tiny (gate stays closed). As V rises, it shoots up 
# linearly, forcing the Na+ gate to OPEN harder and faster the more you depolarize.
def alpha_m(V): return 0.1 * (V + 40.0) / (1.0 - np.exp(-(V + 40.0) / 10.0))

# Exponential Decay (beta_m): The Release.
# At rest (low V), this rate is huge, locking the gate CLOSED. As V rises, 
# this drops to zero instantly, removing the force that holds the gate shut.
def beta_m(V):  return 4.0 * np.exp(-(V + 65.0) / 18.0)

# Exponential Decay (alpha_h): The "Timer" (Inactivation).
# As V rises, this rate drops to zero. It means the "open" probability for 'h' 
# disappears, eventually shutting off the Sodium current even if V stays high.
def alpha_h(V): return 0.07 * np.exp(-(V + 65.0) / 20.0)

# Sigmoid / Logistic (beta_h): The "Refractory Limit."
# This function plateaus. It puts a speed limit on how fast 'h' can recover 
# (re-open) after a spike, ensuring the neuron stays "dead" for a few ms.
def beta_h(V):  return 1.0 / (1.0 + np.exp(-(V + 35.0) / 10.0))

# Boltzmann Function (alpha_n): The "Delayed" Opening.
# Like alpha_m, this opens the gate when V rises, but the constants (0.01 vs 0.1) 
# make it 10x slower. This delay ensures K+ doesn't kill the spike before it starts.
def alpha_n(V): return 0.01 * (V + 55.0) / (1.0 - np.exp(-(V + 55.0) / 10.0))

# Exponential Decay (beta_n): The "Slow Close."
# When V drops back to rest, this rate becomes dominant, slowly closing the K+ channel.
# This sluggish closing is what causes the "undershoot" (hyperpolarization).
def beta_n(V):  return 0.125 * np.exp(-(V + 65.0) / 80.0)
```

```python
def hodgkin_huxley(y, t, I_ext_func):
    V, m, h, n = y
    
    I_inj = I_ext_func(t)
    
    I_Na = g_Na * (m**3) * h * (V - E_Na)
    I_K  = g_K  * (n**4) * (V - E_K)
    I_L  = g_L  * (V - E_L)
    
    dVdt  = (I_inj - I_Na - I_K - I_L) / C_m
    dmdt  = alpha_m(V) * (1 - m) - beta_m(V) * m
    dhdt  = alpha_h(V) * (1 - h) - beta_h(V) * h
    dndt  = alpha_n(V) * (1 - n) - beta_n(V) * n
    
    return [dVdt, dmdt, dhdt, dndt]
```

```python
def run_simulation(time_span, current_func):
    V_rest = -65.0

    m0 = alpha_m(V_rest) / (alpha_m(V_rest) + beta_m(V_rest))
    h0 = alpha_h(V_rest) / (alpha_h(V_rest) + beta_h(V_rest))
    n0 = alpha_n(V_rest) / (alpha_n(V_rest) + beta_n(V_rest))
    
    y0 = [V_rest, m0, h0, n0]
    
    sol = odeint(hodgkin_huxley, y0, time_span, args=(current_func,))
    
    return sol
```

### The Standard Spike & Gating Variables

```python
t = np.linspace(0, 50, 1000)

def I_pulse(t): return 10.0 if 10 <= t <= 11 else 0.0

sol = run_simulation(t, I_pulse)
V, m, h, n = sol.T

fig, ax = plt.subplots(3, 1, figsize=(10, 12), sharex=True)
ax[0].set_title("Scenario A: Anatomy of an Action Potential")
ax[0].plot(t, V, 'k', lw=2)
ax[0].set_ylabel("Voltage (mV)")
ax[0].grid(True)

ax[1].plot(t, m, 'r', label='m (Na activation)')
ax[1].plot(t, h, 'g', label='h (Na inactivation)')
ax[1].plot(t, n, 'b', label='n (K activation)')
ax[1].set_ylabel("Probability")
ax[1].legend()
ax[1].grid(True)
ax[1].text(15, 0.5, "Note 'n' rising slowly (Relaxation)\nand 'h' dropping slowly", fontsize=9)

I_Na = g_Na * (m**3) * h * (V - E_Na)
I_K  = g_K  * (n**4) * (V - E_K)
ax[2].plot(t, I_Na, 'r', label='Na Current')
ax[2].plot(t, I_K, 'b', label='K Current')
ax[2].set_ylabel("Current (uA/cm^2)")
ax[2].set_xlabel("Time (ms)")
ax[2].legend()
ax[2].grid(True)
plt.tight_layout()
```

### Refractory Period (The "Dead" Time)
We fire two pulses. If the second is too close, it fails. When $h=1$, the inactivation gate is open (allowing ions to pass), and when $h=0$, the inactivation gate is closed (blocking ions). Because the formula for conductance is a multiplication ($m^3 \times h$), if $h$ is zero, the entire term becomes zero, effectively blocking the channel regardless of what the $m$ gates are doing.

The refractory period is the recovery time after an action potential during which the neuron is physically unable to trigger a second spike.Time is needed because the sodium inactivation gate ($h$) is slow to reopen (reset from 0 to 1), and until it recovers, the channel remains blocked regardless of how many activation gates ($m$) are open.

```python
t_ref = np.linspace(0, 30, 1000)

def I_double_pulse(t):
    pulse1 = 15.0 if 5 <= t <= 6 else 0.0
    # Try changing pulse 2 to closer towards pulse 1
    pulse2 = 15.0 if 25 <= t <= 26 else 0.0 
    return pulse1 + pulse2

sol_ref = run_simulation(t_ref, I_double_pulse)
V_ref = sol_ref[:, 0]
h_ref = sol_ref[:, 2] # Look at h (inactivation)

fig2, ax2 = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
ax2[0].set_title("Scenario B: Relative Refractory Period")
ax2[0].plot(t_ref, V_ref, 'k')
ax2[0].set_ylabel("Voltage (mV)")
ax2[0].grid(True)

# Why did it struggle? Look at 'h'
ax2[1].plot(t_ref, h_ref, 'g', label='h (Availability)')
ax2[1].axhline(0.5, color='gray', linestyle='--')
ax2[1].set_ylabel("h Gate")
ax2[1].set_xlabel("Time (ms)")
ax2[1].legend()
ax2[1].text(16, 0.2, "h is still low!\nNa channels are 'tired'", color='green')
plt.tight_layout()
```

### Anode Break Excitation (The "Rebound")

At normal resting potential, about 40% of your sodium channels are chemically blocked (inactivated), meaning the $h$ gate is only partially open.When you hyperpolarize the cell (force the voltage down), you physically remove this block, resetting the $h$ gate to be nearly 100% open and making the channels "super-primed."

When you release the voltage, it naturally swings back up to rest, but because the channels are now fully available instead of partially blocked, this simple return swing is enough to trigger a spontaneous spike.

```python
t_reb = np.linspace(0, 50, 1000)

def I_hyper(t): return -10.0 if 10 <= t <= 30 else 0.0 # Negative current

sol_reb = run_simulation(t_reb, I_hyper)
V_reb = sol_reb[:, 0]

plt.figure(figsize=(8, 4))
plt.title("Scenario C: Anode Break (Rebound Spike)")
plt.plot(t_reb, V_reb, 'k')
plt.ylabel("Voltage (mV)")
plt.xlabel("Time (ms)")
plt.grid(True)
plt.text(32, -20, "Spike upon release!", color='red')
plt.text(15, -80, "Hyperpolarization removes\nNa inactivation (resets 'h')", ha='center')
```

### Phase Space

```python
plt.figure(figsize=(6, 6))
plt.title("Scenario D: Phase Space (V vs n)")
plt.plot(V, n, 'b-', label='Trajectory')
plt.plot(V[0], n[0], 'ro', label='Rest') # Start
plt.xlabel("Membrane Potential V (mV)")
plt.ylabel("Potassium Gate n")
plt.grid(True)
plt.legend()

plt.show()
```

The phase space plot (Voltage $V$ vs. Potassium Gate $n$) visualizes the cycle of the action potential as a loop rather than a flat line.

1. The Start: It begins at rest (low $V$, low $n$).
2. The Spike (Right): The voltage $V$ shoots up rapidly (moving right), but $n$ stays low because the potassium gates are slow (delayed).
3. The Recovery (Top): Once $V$ is high, the potassium gates finally catch up and open ($n$ goes up), which forces the voltage back down.
4. The Reset (Left/Bottom): The voltage drops below rest (hyperpolarization) while $n$ slowly closes, eventually bringing the system back to the starting point.

In short the loop shape proves that $n$ (potassium) is always chasing $V$ (voltage) but is always lagging behind it. At the very end of the loop (bottom left), the voltage is lower than the resting potential (hyperpolarized) because the potassium gates ($n$) are still stuck open.Because the voltage is so low, it slowly forces these $n$ gates to close, shutting off the potassium outflow.Once they finally close, the potassium current stops fighting, and the background "leak" channels naturally pull the voltage back up to the starting resting point.
