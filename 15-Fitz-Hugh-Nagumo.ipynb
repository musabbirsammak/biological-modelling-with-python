## 1. The Biological Foundation: The Neuron

Before diving into the calculus, we must understand the "characters" in our story. A **neuron** is a specialized cell designed for rapid communication.

- **Dendrites:** These are the input fibers that receive information.
- **The Axon:** This is the output cable that transports information via electrical impulses called **action potentials**.
- **The Threshold:** Neurons are binary in their logic. A mild stimulus does nothing; however, once a stimulus crosses a specific **threshold**, it triggers a full-scale electrical surge.
## 2. From Hodgkin-Huxley to Fitzhugh-Nagumo

In 1952, Hodgkin and Huxley modeled this using four complex differential equations. Fitzhugh and Nagumo simplified this into a two-variable system by noting that **sodium channels ($Na^{+}$)** react much faster than **potassium channels ($K^{+}$)**. The resulting system is:
$$\frac{dV}{dt} = -V(V - a)(V - 1) - w$$
$$\frac{dw}{dt} = b(V - cw)$$
### Rationalizing the Variables

- **$V$ (Potential):** Represents the voltage across the cell membrane.
- **$w$ (Recovery):** A recovery variable modeling the slower ion channel dynamics.
- **$a$ (Threshold):** This constant ($0 < a < 1$) is the tripwire. If $V$ doesn't reach $a$, the signal dies.
- **$b, c$ (Scaling constants):** Positive parameters that govern the speed and steady-state relationship of the recovery variable.
## 3. The Derivation of Stability

### Step A: Finding the Jacobian Matrix

To understand how the system behaves near the "resting state" (0,0), we use a **Jacobian Matrix ($Df$)**. We use this tool because it "linearizes" a complex curve into a flat plane at a specific point, allowing us to see which way the "wind" of the differential equations is blowing. The general form is:
$$Df(V, w) = \begin{bmatrix} \frac{\partial f_1}{\partial V} & \frac{\partial f_1}{\partial w} \\ \frac{\partial f_2}{\partial V} & \frac{\partial f_2}{\partial w} \end{bmatrix}$$
1. **Differentiate $f_1 = -V(V-a)(V-1) - w$:**    
    - Expand $f_1$: $-V(V^2 - V - aV + a) - w = -V^3 + (1+a)V^2 - aV - w$.
    - $\frac{\partial f_1}{\partial V} = -3V^2 + 2(1+a)V - a$.
    - $\frac{\partial f_1}{\partial w} = -1$.
2. **Differentiate $f_2 = b(V - cw)$:**    
    - $\frac{\partial f_2}{\partial V} = b$.
    - $\frac{\partial f_2}{\partial w} = -bc$.

At the equilibrium **$(0, 0)$**, the matrix simplifies to:
$$Df(0, 0) = \begin{bmatrix} -a & -1 \\ b & -bc \end{bmatrix}$$
### Step B: The Characteristic Equation

To find the **eigenvalues ($\lambda$)**, which tell us if the system returns to zero (stable) or explodes (unstable), we solve $det(A - \lambda I) = 0$:
$$(-a - \lambda)(-bc - \lambda) - (-1)(b) = 0$$
**The expansion:**

1. Multiply the terms: $abc + a\lambda + bc\lambda + \lambda^2 + b = 0$.    
2. Group by powers of $\lambda$: $\lambda^2 + (a + bc)\lambda + (abc + b) = 0$.
3. Factor out $b$: $\lambda^2 + (a + bc)\lambda + b(ac + 1) = 0$.
### Step C: Solving for $\lambda$ (The Quadratic Formula)

Using $\lambda = \frac{-B \pm \sqrt{B^2 - 4AC}}{2}$:
$$\lambda = \frac{-(a + bc) \pm \sqrt{(a + bc)^2 - 4b(ac + 1)}}{2}$$
Expand the discriminant (the part inside the square root):
- $(a + bc)^2 - 4b(ac + 1) = a^2 + 2abc + b^2c^2 - 4abc - 4b$
Combine like terms ($2abc - 4abc = -2abc$):
- $a^2 - 2abc + b^2c^2 - 4b$
Recognize the perfect square $(a - bc)^2$:

**Discriminant** $= (a - bc)^2 - 4b$
## 4. Why is it Stable? (The Narrative Logic)

For the resting state to be stable, the real parts of these eigenvalues must be **negative**.

1. **The Real Part:** Look at the leading term: $-(a + bc)$. Since $a, b, c$ are all positive, this term is always negative.
2. **The "Pull" of the Square Root:** The square root term $\sqrt{(a+bc)^2 - 4b(ac+1)}$ is mathematically guaranteed to be smaller than the absolute value of the leading term $(a+bc)$ because we are subtracting a positive number ($4b(ac+1)$) from it.
3. **The Result:** Even when you add the square root, the numerator remains negative.

Imagine a ball in a bowl. No matter which way you nudge it, the "negative" eigenvalues act like gravity, pulling the ball back to the center (the resting state).
## 5. The Action Potential: Crossing the Rubicon

The "magic" of the Fitzhugh-Nagumo model is what happens when the stimulus is **strong**.

- **Weak Stimulus ($V < a$):** The cubic function's internal "braking" is too strong. The trajectory makes a small loop and returns to $(0,0)$.
- **Strong Stimulus ($V > a$):** The $V$ variable enters a "positive feedback" zone. The cubic geometry of the $dV/dt$ equation forces the trajectory to take a massive detour—swinging far out into the phase plane before the slower $w$ variable (the recovery) can eventually pull it back to zero.

This massive detour is the **Action Potential**.
# Dimensionless Version

While the core logic of the Fitzhugh-Nagumo (FHN) model remains the same—using a fast voltage variable and a slow recovery variable—the mathematical form can vary depending on whether the author is using a **phenomenological** version or a **dimensionless** version.
## 1. Comparing the "Two Worlds"

Standard dimensionless form of the FHN equations:
$$\dot{V} = V - V^3 - W + I$$
$$\dot{W} = \epsilon(V - aW + b)$$
### Rationalizing the New Symbols

- **$-V^3$:** In previous version, the cubic nature came from the term $-V(V-a)(V-1)$. Here, it is explicitly simplified to $V - V^3$. This provides the "N-shaped" curve necessary for the threshold effect.
- **$I$ (Injected Current):** This represents an external stimulus. In the previous version, the stimulus was modeled as an initial change in $V$. Here, $I$ is a constant "push" that can shift the entire system.
- **$\epsilon$ (Small Parameter):** This represents the **time-scale separation**. Since $\epsilon$ is small (e.g., $0.1$), the change in $W$ ($\dot{W}$) is much slower than the change in $V$. This mimics how potassium channels open more slowly than sodium channels.
## 2. Deriving the Nullclines (Part A)

To find the "geography" of the system, we set the rates of change to zero.
### The V-nullcline ($\dot{V} = 0$)

1. **Start with:** $0 = V - V^3 - W + I$.    
2. **Isolate $W$:** Add $W$ to both sides.
3. **Result:** $W = V - V^3 + I$.
    - **Visual Anchor:** This is a cubic "hill and valley" curve. At $I=0$, it passes through the origin. If you increase $I$, the entire "hill" lifts upward.
### The W-nullcline ($\dot{W} = 0$)

1. **Start with:** $0 = \epsilon(V - aW + b)$.
2. **Divide by $\epsilon$:** $0 = V - aW + b$.
3. **Isolate $W$:** $aW = V + b$.
4. **Result:** $W = \frac{1}{a}V + \frac{b}{a}$.
    - **Visual Anchor:** This is a straight line. The intersections between this line and the cubic "hill" are your **steady states** (equilibria).
## 3. Finding the "Resting State" (Part B)

If we set $a = 0.5$, $b = 1$, and $I = 0$, we can find the exact coordinates of the resting neuron.

**The Equations to Solve:**

1. $W = V - V^3$
2. $W = \frac{V + 1}{0.5} \rightarrow W = 2V + 2$

**The Derivation:**

- Substitute $W$: $2V + 2 = V - V^3$.
- Rearrange into a polynomial: $V^3 + V + 2 = 0$.
- **The "No-Leaps" Factorization:** By testing small integers, we see that $V = -1$ works: $(-1)^3 + (-1) + 2 = -1 - 1 + 2 = 0$.
- Using synthetic division or grouping: $(V+1)(V^2 - V + 2) = 0$.
- The quadratic part ($V^2 - V + 2$) has no real roots (the discriminant $1 - 8$ is negative).
- **Result:** $V = -1$.
- Substitute back to find $W$: $W = 2(-1) + 2 = 0$.

The **Resting State** is at **$(-1, 0)$**.
## 4. The Mathematical "Flipping Point"

The project asks what happens when you increase $I$ beyond $1.3$.
### The Hopf Bifurcation

As you increase $I$, the cubic nullcline shifts up. This moves the intersection point (the steady state) from the stable left branch of the cubic curve to the middle section, where the slope is positive.

**Why the behavior changes:**

- **Below $I \approx 1.3$:** The steady state is a "stable sink". The neuron receives a push but always returns to its resting voltage.
- **At $I \approx 1.3$:** A **Hopf Bifurcation** occurs. The steady state becomes unstable.
- **Above $I \approx 1.3$:** The system can no longer stay at a single point. Instead, it gets trapped in a **Limit Cycle**.

**Mental Model:** The neuron stops being a "receiver" and becomes an "oscillator". Instead of firing once and stopping, it fires a continuous train of spikes (repetitive firing) because the injected current $I$ is so strong it never lets the voltage rest.

```python
import numpy as np
import numpy.typing as npt
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def fhn_model(y: npt.NDArray, t: float, a: float, b: float,
              eps: float, I: float) -> npt.NDArray:
    """
    The Fitz-Hugh Nagumo Model.

    Parameters
    ==========
    y: npt.NDArray
        The state vector.y[0] is `V` or the membrane potential.
        y[1] is `W` or the recovery variable.
    t: float
        Time.
    a: float
        Parameter that control the resting state.
    b: float
        Parameter that control the firing.
    eps: float
        Timescale difference between fast V and slow W.
    I: float
        Input current.
    Returns
    =======
    list[float]
        Time derivative of V and W.
    """
    V, W = y
    
    dVdt = V - V**3 - W + I
    dWdt = eps * (V - a*W + b)
    
    return np.array([dVdt, dWdt])
```

## Nullclines and Steady States

```python
a = 12
b = 1
I = 0.0

V_range = np.linspace(-2.5, 2.5, 400)

W_null_V = V_range - V_range**3 + I
W_null_W = (V_range + b) / a

plt.figure(figsize=(10, 7))
plt.plot(V_range, W_null_V, 'b-', lw=2, label='V-Nullcline')
plt.plot(V_range, W_null_W, 'r-', lw=2, label='W-Nullcline')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("Nullclines in the VW-plane")
plt.xlabel("Membrane Voltage (V)")
plt.ylabel("Recovery Variable (W)")
plt.ylim(-2, 2)
plt.xlim(-2.5, 2.5)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
a = 0.5
b = 1.0
I = 0

V_range = np.linspace(-2.5, 2.5, 400)

W_null_V = V_range - V_range**3 + I
W_null_W = (V_range + b) / a

plt.figure(figsize=(10, 7))

plt.plot(V_range, W_null_V, 'b-', lw=2, label='V-Nullcline')
plt.plot(V_range, W_null_W, 'r-', lw=2, label='W-Nullcline')
plt.plot(-1, 0, 'go', markersize=10, label='Resting State (-1, 0)')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("Nullclines in the VW-plane")
plt.xlabel("Membrane Voltage (V)")
plt.ylabel("Recovery Variable (W)")
plt.ylim(-2, 2)
plt.xlim(-2.5, 2.5)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.savefig("./resting-state.pdf")
plt.show()
```

## Trajectories

```python
a = 0.5
b = 1.0
eps = 0.1
I = 0

T = 1000
nt = 10000
t = np.linspace(0, T, nt)

initial_conds = [[0, 0], [-1, 0.5], [-1.75, -0.75]]

plt.figure(figsize=(12, 6))

for cond in initial_conds:
    sol = odeint(fhn_model, cond, t, args=(a, b, eps, I))
    plt.plot(t, sol[:, 0], label=f"{cond[0]},{cond[1]}")

plt.title("Membrane Voltage V over Time")
plt.xlabel("Time (t)")
plt.ylabel("Voltage (V)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.savefig("./voltage-time.pdf")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

V_range = np.linspace(-2.5, 2.5, 400)
W_null_V = V_range - V_range**3 + I
W_null_W = (V_range + b) / a

plt.plot(V_range, W_null_V, 'k--', alpha=0.3, label='Nullclines')
plt.plot(V_range, W_null_W, 'k--', alpha=0.3)

colors = ['r', 'g', 'b']
for i, cond in enumerate(initial_conds):
    sol = odeint(fhn_model, cond, t, args=(a, b, eps, I))
    plt.plot(sol[:, 0], sol[:, 1], color=colors[i], label=f'{cond}')
    plt.plot(cond[0], cond[1], 'o', color=colors[i])

plt.plot(-1, 0, 'ko', markersize=8)
plt.text(-1, 0.1, " Resting State (-1, 0)")

plt.title("Trajectories in Phase Space (V-W Plane)")
plt.xlabel("Voltage (V)")
plt.ylabel("Recovery Variable (W)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xlim(-2, 1.5)
plt.ylim(-1.5, 1.5)

plt.savefig("phase-space.pdf")
plt.show()
```

## Impact of I

```python
a = 0.5
b = 1.0
I = [0, 0.5, 1, 1.5]

V_range = np.linspace(-2.5, 2.5, 400)
plt.figure(figsize=(10, 7))

for count, i in enumerate(I):
    W_null_V = V_range - V_range**3 + i
    W_null_W = (V_range + b) / a

    if count == 0:
        plt.plot(V_range, W_null_W, '-', color="navy", lw=3, label='W-Nullcline')
    
    plt.plot(V_range, W_null_V, '--', lw=2, label=f'V-Nullcline I={i}')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("Impact of Different I on Nullclines and Steady States")
plt.xlabel("Membrane Voltage (V)")
plt.ylabel("Recovery Variable (W)")
plt.ylim(-2, 2)
plt.xlim(-2.5, 2.5)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Limit Cycles

```python
a, b, eps = 0.5, 1.0, 0.1
T, nt = 1000, 1000
t = np.linspace(0, T, nt)
y0 = [0, 0]

# Values of I to test
I_values = [0.5, 1.1, 1.25, 1.3, 1.35, 1.5, 2.5, 4]

# Create Subplots
fig, axes = plt.subplots(4, 2, figsize=(14, 10))
axes = axes.flatten()

for i, I_val in enumerate(I_values):
    sol = odeint(fhn_model, y0, t, args=(a, b, eps, I_val))
    V = sol[:, 0]
    
    axes[i].plot(t, V, color='tab:red')
    axes[i].set_title(f"Injected Current $I = {I_val}$")
    axes[i].set_xlabel("Time ($t$)")
    axes[i].set_ylabel("Voltage ($V$)")
    axes[i].grid(True, alpha=0.3)
    axes[i].set_ylim(-2, 2)

plt.tight_layout()
plt.suptitle("Transition from Resting State to Repetitive Firing (Hopf Bifurcation)", y=1.02, fontsize=16)
plt.show()
```
