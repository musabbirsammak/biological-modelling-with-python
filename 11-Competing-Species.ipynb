## Competing Species

The model presented in the slides is a system of two coupled, non-linear ordinary differential equations (ODEs). It extends the standard single-species Logistic Growth model to two interacting populations.

The model considers two species—specifically illustrated as **cows** ($x_1$) and **goats** ($x_2$)—occupying the same environment. The system is built by modifying the **Logistic Model** for each species to include a "competition term":
$$\begin{cases} \dot{x}_{1}=r_{1}x_{1}\frac{B_{1}-x_{1}-\gamma_{1}x_{2}}{B_{1}} \\ \dot{x}_{2}=r_{2}x_{2}\frac{B_{2}-x_{2}-\gamma_{2}x_{1}}{B_{2}} \end{cases}$$
Before solving, we must rigorously define every symbol to understand the "physical" logic of the equation.

- **$x_1, x_2$ (Population Size):** The number of individuals of cows ($x_1$) and goats ($x_2$) at time $t$.
    - _Unit:_ Number of individuals.
- **$r_1, r_2$ (Intrinsic Growth Rate):** The maximum per capita growth rate when resources are unlimited and no competitors are present.
    - _Unit:_ $1/time$ (e.g., per year).
- **$B_1, B_2$ (Carrying Capacity):** The maximum population size species $x_i$ can support in isolation. In many textbooks this is denoted as $K$, but here it is $B$.
    - _Unit:_ Number of individuals.
- **The Term $(B_1 - x_1)$:** This represents the "remaining free space" or available resources for species 1 within its own niche.
- **$\gamma_1, \gamma_2$ (Competition Coefficients):** These are the critical coupling parameters.
    - $\gamma_1$: The equivalent number of cows that one goat displaces. It quantifies the effect of the presence of one goat on the cow population.
    - If $\gamma_1 > 1$, a goat consumes more resources than a cow (strong competition).
    - If $\gamma_1 < 1$, a goat consumes fewer resources than a cow (weak competition).

To justify this mathematical structure, we must accept the following biological premises:

- **Limited Resources:** Neither population can grow indefinitely; they are bounded by a carrying capacity ($B$).
- **Logistic Growth:** In the absence of the competitor (e.g., if $x_2=0$), species $x_1$ follows a standard logistic S-curve.
- **Linear Interaction:** The negative effect of one species on the other is proportional to the population size (density-dependent competition).
- **Constant Environment:** Parameters like growth rate ($r$) and carrying capacity ($B$) do not change with time.

For coupled non-linear systems like this, an explicit algebraic solution $x(t)$ (a formula for population over time) is analytically intractable. Instead, we "solve" the system by analyzing its **Phase Plane Dynamics**. We determine where the system moves and where it stops.

**The $x_1$-Nullcline (Cows are stationary):** 
We set the rate of change for cows to zero:
$$\frac{dx_1}{dt} = 0$$
Substitute the full equation:
$$r_{1}x_{1}\left(\frac{B_{1}-x_{1}-\gamma_{1}x_{2}}{B_{1}}\right) = 0$$
For this product to be zero, one of the factors must be zero.

- **Factor 1:** $x_1 = 0$. (This is the y-axis; if there are no cows, the population stays at zero).
- **Factor 2:** The term inside the parenthesis is zero.$$B_{1} - x_{1} - \gamma_{1}x_{2} = 0$$
We rearrange this to slope-intercept form ($x_2 = mx_1 + c$) to plot it easily on the $x_1-x_2$ plane. $$B_1 - x_1 = \gamma_1 x_2$$Divide by $\gamma_1$:
$$\frac{B_1}{\gamma_1} - \frac{x_1}{\gamma_1} = x_2$$
So, the non-trivial $x_1$-nullcline is the line:
$$x_2 = -\frac{1}{\gamma_1}x_1 + \frac{B_1}{\gamma_1}$$
Anywhere along this line, the cow population is constant. Below it, cows increase; above it, they decrease.    

**The $x_2$-Nullcline (Goats are stationary):**
We set the rate of change for goats to zero:
$$\frac{dx_2}{dt} = 0$$
Substitute the equation: $$r_{2}x_{2}\left(\frac{B_{2}-x_{2}-\gamma_{2}x_{1}}{B_{2}}\right) = 0$$Again, two solutions:

- **Factor 1:** $x_2 = 0$. (The x-axis).
- **Factor 2:**$$B_{2} - x_{2} - \gamma_{2}x_{1} = 0$$Rearrange for $x_2$:
$$x_2 = -\gamma_2 x_1 + B_2$$
So, the non-trivial $x_2$-nullcline is the line:
$$x_2 = B_2 - \gamma_2 x_1$$

Anywhere along this line, the goat population is constant. Imagine the phase plane (a graph with Cows $x_1$ on the X-axis and Goats $x_2$ on the Y-axis).

- **The "Push":** The growth rates $r$ push the populations upward and rightward (growth).
- **The "Wall":** The Nullclines act as invisible barriers.
    - The **Cow Nullcline** hits the x-axis at $B_1$ (carrying capacity) and the y-axis at $B_1/\gamma_1$ **(the number of goats needed to eat all the cow food)**.
    - The **Goat Nullcline** hits the y-axis at $B_2$ and the x-axis at $B_2/\gamma_2$. **(the number of cows needed to eat all the goat food)**.

The behavior of the system depends entirely on how these two lines cross each other. We determine the long-term outcome by checking where the nullclines intersect. These intersections are the **Fixed Points (Equilibria)**.

Case 1: Small Interaction (Weak Competition)

- **Condition:** $\gamma_1$ and $\gamma_2$ are small.
- **Geometric Consequence:**
    - The Cow Nullcline is steeper than the Goat Nullcline.
    - $B_1 < B_2/\gamma_2$ (Cows can't out-eat Goats easily).
    - $B_2 < B_1/\gamma_1$ (Goats can't out-eat Cows easily).

To understand why "small interactions" ($\gamma < 1$) lead to those specific geometric consequences and steepness, we have to look strictly at the algebra of the lines plotted on the phase plane. We are comparing two straight lines. To understand their geometry (steepness and position), we simply need to calculate their Slopes and Intercepts derived from the nullcline equations. Let's rewrite the nullclines in the standard slope-intercept form ($y = mx + q$), treating $x_2$ as the $y$-axis and $x_1$ as the $x$-axis.

**The Cow Nullcline ($x_1$-nullcline):**
$$x_2 = \underbrace{-\frac{1}{\gamma_1}}_{\text{Slope } m_1} x_1 + \underbrace{\frac{B_1}{\gamma_1}}_{\text{Y-Intercept}}$$
- **Slope:** $-\frac{1}{\gamma_1}$
- **Y-Intercept:** $B_1/\gamma_1$
- **X-Intercept:** $B_1$ (Set $x_2=0$)

**The Goat Nullcline ($x_2$-nullcline):**
$$x_2 = \underbrace{-\gamma_2}_{\text{Slope } m_2} x_1 + \underbrace{B_2}_{\text{Y-Intercept}}$$

- **Slope:** $-\gamma_2$
- **Y-Intercept:** $B_2$
- **X-Intercept:** $B_2/\gamma_2$ (Set $x_2=0 \rightarrow \gamma_2 x_1 = B_2$)

Mathematically, "steepness" refers to the magnitude of the slope $|m|$.

- **Cow Slope magnitude:** $|m_1| = \frac{1}{\gamma_1}$    
- **Goat Slope magnitude:** $|m_2| = \gamma_2$

In this scenario, we assume weak competition, meaning $\gamma_1 < 1$ and $\gamma_2 < 1$ (a single individual consumes less of the _other_ species' resources than its own).

- If $\gamma_1$ is a small fraction (e.g., $0.1$), then $\frac{1}{\gamma_1}$ becomes a **large number** (e.g., $10$). $\rightarrow$ **Very Steep Line.**    
- If $\gamma_2$ is a small fraction (e.g., $0.1$), it remains a **small number**. $\rightarrow$ **Very Flat (Shallow) Line.**

The Cow nullcline is nearly vertical (steep), and the Goat nullcline is nearly horizontal (flat). This is purely a result of $\gamma$ being in the denominator for Cows and the numerator for Goats.

The inequalities determine if one line lies completely "inside" or "outside" the other, or if they cross. Let's analyze the **X-Axis Intercepts** (where population $x_2 = 0$):

1. **Cow X-Intercept:** $B_1$ (This is the carrying capacity of Cows).
2. **Goat X-Intercept:** $B_2/\gamma_2$ (This is the amount of Cows needed to consume all Goat resources).

**The Inequality:** $B_1 < B_2/\gamma_2$

- **Geometric Translation:** On the horizontal x-axis, the Cow line hits the axis **to the left** of the Goat line.
- **Biological Meaning:** Even if the Cow population reaches its maximum ($B_1$), it is _not enough_ to consume all the resources Goats need ($B_2/\gamma_2$). Therefore, Cows cannot drive Goats extinct purely by resource depletion.

Now let's analyze the **Y-Axis Intercepts** (where population $x_1 = 0$):

1. **Cow Y-Intercept:** $B_1/\gamma_1$ (The amount of Goats needed to consume all Cow resources).    
2. **Goat Y-Intercept:** $B_2$ (The carrying capacity of Goats).

**The Inequality:** $B_2 < B_1/\gamma_1$

- **Geometric Translation:** On the vertical y-axis, the Goat line hits the axis **below** the Cow line.
- **Biological Meaning:** Even if the Goat population reaches its maximum ($B_2$), it is _not enough_ to consume all the resources Cows need ($B_1/\gamma_1$). Goats cannot drive Cows extinct.

Because of these two inequalities, we have a geometric contradiction if we try to avoid an intersection:

1. On the **Y-Axis** (Left), the Cow line is **Higher** ($B_1/\gamma_1 > B_2$).
2. On the **X-Axis** (Right), the Cow line is **Lower** ($B_1 < B_2/\gamma_2$).

If a straight line starts above another line but ends below it, it must cross it. This crossing point is the **Stable Equilibrium** where the two species coexist.

- **Dynamics:** The lines intersect at a point $(x_1^*, x_2^*)$ in the middle of the positive quadrant.
- **Stability:** Because competition is weak ($\gamma$ is small), each species inhibits its own growth more than it inhibits the other. The system stabilizes at this intersection.
- **Result:** **Stable Coexistence.** Both cows and goats survive.

Case 2: Large Interaction (Strong Competition)

- **Condition:** $\gamma_1$ and $\gamma_2$ are large.
- **Geometric Consequence:**
    - $B_1 > B_2/\gamma_2$ and $B_2 > B_1/\gamma_1$.
- **Dynamics:** The lines intersect, but the flow vectors point away from the intersection.
- **Stability:** The coexistence point becomes an unstable saddle point. A small perturbation pushes the system toward one of the axes.
- **Result:** **Competitive Exclusion.** Depending on who had the head start, one species drives the other to extinction.

To prove stability mathematically, we calculate the Jacobian Matrix $J$ at an equilibrium point $(x_1, x_2)$.
$$J(x_1, x_2) = \begin{pmatrix} \frac{\partial \dot{x}_1}{\partial x_1} & \frac{\partial \dot{x}_1}{\partial x_2} \\ \frac{\partial \dot{x}_2}{\partial x_1} & \frac{\partial \dot{x}_2}{\partial x_2} \end{pmatrix}$$
Let's expand the derivative for $\dot{x}_1$:

$$\dot{x}_1 = r_1 x_1 - \frac{r_1}{B_1}x_1^2 - \frac{r_1 \gamma_1}{B_1}x_1 x_2$$
The partial derivatives are:
- $\frac{\partial \dot{x}_1}{\partial x_1} = r_1 - \frac{2r_1 x_1}{B_1} - \frac{r_1 \gamma_1 x_2}{B_1}$
- $\frac{\partial \dot{x}_1}{\partial x_2} = - \frac{r_1 \gamma_1 x_1}{B_1}$

Similarly for $\dot{x}_2$:
- $\frac{\partial \dot{x}_2}{\partial x_1} = - \frac{r_2 \gamma_2 x_2}{B_2}$
- $\frac{\partial \dot{x}_2}{\partial x_2} = r_2 - \frac{2r_2 x_2}{B_2} - \frac{r_2 \gamma_2 x_1}{B_2}$

**Stability Criteria:**

We evaluate $J$ at the equilibrium point.

1. Calculate the **Trace** ($Tr = \lambda_1 + \lambda_2$) and **Determinant** ($Det = \lambda_1 \lambda_2$).
2. **Stable Sink:** If $Tr < 0$ and $Det > 0$, the species coexist (Case 1).
3. **Saddle Point:** If $Det < 0$, the coexistence is unstable (Case 2).
 
 In the single-species logistic model, a population always settles at its carrying capacity $B$. In this competing system, species $x_1$ can never reach $B_1$ as long as species $x_2$ is present, because the term $-\gamma_1 x_2$ effectively reduces the "available" carrying capacity.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def competition_model(state, t, r1, r2, B1, B2, g1, g2):
    x1, x2 = state
    
    dx1dt = r1 * x1 * ((B1 - x1 - g1 * x2) / B1)
    dx2dt = r2 * x2 * ((B2 - x2 - g2 * x1) / B2)
    
    return [dx1dt, dx2dt]
```

```python
def plot_phase_plane(ax, r1, r2, B1, B2, g1, g2, title):
    # Grid for vector field
    x1_max = max(B1, B2/g2) * 1.2
    x2_max = max(B2, B1/g1) * 1.2
    
    x1_vals = np.linspace(0, x1_max, 20)
    x2_vals = np.linspace(0, x2_max, 20)
    X1, X2 = np.meshgrid(x1_vals, x2_vals)
    
    # Calculate flow vectors (dX/dt) at every grid point
    U = r1 * X1 * ((B1 - X1 - g1 * X2) / B1)
    V = r2 * X2 * ((B2 - X2 - g2 * X1) / B2)
    
    # Plot Streamlines (The "Flow" of the system)
    ax.streamplot(X1, X2, U, V, density=1.5, color='lightgray', arrowsize=1.5)
    
    # --- Plot Nullclines ---
    # Cow Nullcline (x1-nullcline): dx1/dt = 0 -> x2 = (B1 - x1)/g1
    x1_line = np.linspace(0, B1, 100)
    x2_cow_null = (B1 - x1_line) / g1
    ax.plot(x1_line, x2_cow_null, 'b-', lw=3, label=f'Cow Nullcline (dx1/dt=0)')
    
    # Goat Nullcline (x2-nullcline): dx2/dt = 0 -> x2 = B2 - g2*x1
    # We plot x2 against x1. The line hits x-axis at B2/g2.
    x1_line_goat = np.linspace(0, B2/g2, 100)
    x2_goat_null = B2 - g2 * x1_line_goat
    ax.plot(x1_line_goat, x2_goat_null, 'r-', lw=3, label=f'Goat Nullcline (dx2/dt=0)')
    
    # --- Plot Trajectories ---
    # Start from random points to see where they end up
    t = np.linspace(0, 50, 1000)
    initial_conditions = [(B1*0.1, B2*0.1), (B1*0.8, B2*1.2), (B1*1.2, B2*0.1)]
    
    for x0 in initial_conditions:
        sol = odeint(competition_model, x0, t, args=(r1, r2, B1, B2, g1, g2))
        ax.plot(sol[:, 0], sol[:, 1], 'k--', lw=1.5) # Path
        ax.plot(sol[0, 0], sol[0, 1], 'ko') # Start
        ax.plot(sol[-1, 0], sol[-1, 1], 'k*', markersize=10) # End (Equilibrium)

    # Formatting
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('Cows ($x_1$)')
    ax.set_ylabel('Goats ($x_2$)')
    ax.set_xlim(0, x1_max)
    ax.set_ylim(0, x2_max)
    ax.legend(loc='upper right')
    ax.grid(alpha=0.3)
```

```python
# Common Parameters
r1, r2 = 0.5, 0.5   # Growth rates
B1, B2 = 100, 100   # Carrying capacities

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

g1_weak, g2_weak = 0.5, 0.5 
plot_phase_plane(ax1, r1, r2, B1, B2, g1_weak, g2_weak, 
                 "Case 1: Stable Coexistence\n($\\gamma < 1$: Weak Competition)")

g1_strong, g2_strong = 1.5, 2.0
plot_phase_plane(ax2, r1, r2, B1, B2, g1_strong, g2_strong, 
                 "Case 2: Competitive Exclusion\n($\\gamma > 1$: Strong Competition)")

plt.tight_layout()
plt.show()
```
